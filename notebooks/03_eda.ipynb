# Step 03: Exploratory Data Analysis (EDA)
Goal: Visualize distributions and identify patterns in the cleaned dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

INPUT_FILE = "../data/interim/02_cleaning.csv"
df = pd.read_csv(INPUT_FILE)

sns.set_theme(style="whitegrid")
print("Loaded dataset for EDA")
```

### 1. Loan Amount Distribution

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Loan_Amount_INR'], bins=30, kde=True, color='teal')
plt.title('Distribution of Loan Amount (INR)')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.show()
```

### 2. TAT by Sector

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='Sector', y='Total_TAT_Days', data=df, palette='viridis')
plt.title('Total Turnaround Time (TAT) by Sector')
plt.xticks(rotation=45)
plt.show()
```

### 3. Risk Score Correlation

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Risk_Score', y='Credit_Eligibility_Score', hue='Repayment_Status', data=df)
plt.title('Risk Score vs. Credit Eligibility Score')
plt.show()
```
