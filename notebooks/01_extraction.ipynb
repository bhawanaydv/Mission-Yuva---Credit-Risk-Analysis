# Step 01: Data Extraction
Goal: Load the raw dataset and select relevant features (including Application Date for forecasting).

```python
import pandas as pd
import os

INPUT_FILE = "../data/raw/mission_yuva_credit_flow.csv"
OUTPUT_FILE = "../data/interim/01_extraction.csv"

# Ensure interim directory exists
os.makedirs("../data/interim", exist_ok=True)

df = pd.read_csv(INPUT_FILE)
print(f"Loaded dataset: {df.shape[0]} rows × {df.shape[1]} columns")
```

```python
cols = [
    'Application ID',
    'Application Date',
    'District Name',
    'Gender',
    'Residential Type',
    'Specially Abled Type',
    'Enterprise Type',
    'Sector',
    'IFSC Code',
    'Days_to_BHD_Verification',
    'Days_to_SBDU_Verification',
    'Days_to_DLIC_Approval',
    'Days_to_Bank_Sanction',
    'Days_to_Disbursement',
    'Total_TAT_Days',
    'Loan Amount (Mixed Units)',
    'Project Cost (Mixed Units)',
    'Applicant_Monthly_Income',
    'Loan_Tenure_Months',
    'Interest_Rate_%',
    'EMI_Amount',
    'EMI_to_Income_Ratio',
    'Risk_Score',
    'Credit_Eligibility_Score',
    'Repayment_Status'
]

df = df[cols].copy()
print(f"Columns reduced to {len(cols)}")
```

```python
df.to_csv(OUTPUT_FILE, index=False)
print(f"Saved to {OUTPUT_FILE}")
```
